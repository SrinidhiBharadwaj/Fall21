```python
import skgeom as sg
from skgeom.draw import draw
import matplotlib
import numpy as np
import networkx as nx
import numpy.linalg as LA
import time
from scipy.spatial import cKDTree
```

```python
matplotlib.rcParams['figure.figsize'] = (20, 10)
w1 = sg.Polygon([sg.Point2(800,0),  sg.Point2(0,0)])
w2 = sg.Polygon([sg.Point2(0,0), sg.Point2(0,300)])
w3 = sg.Polygon([sg.Point2(0,300), sg.Point2(800,300)])
w4 = sg.Polygon([sg.Point2(800,300), sg.Point2(800,0)])
w5 = sg.Polygon([sg.Point2(200,0), sg.Point2(200,200)])
w6 = sg.Polygon([sg.Point2(100,200), sg.Point2(300,200)])
w7 = sg.Polygon([sg.Point2(400,300), sg.Point2(400,100)])
w8 = sg.Polygon([sg.Point2(600,0), sg.Point2(600,200)])
w9 = sg.Polygon([sg.Point2(500,200), sg.Point2(700,200)])

env = []
env.append(w1)
env.append(w2)
env.append(w3)
env.append(w4)
env.append(w5)
env.append(w6)
env.append(w7)
env.append(w8)
env.append(w9)

draw(env)
```

```python
def robotGenerator(angle = 0):
    points = np.array([[-25, -25],
              [ 25, -25],
              [ 25,  25],
              [-25,  25]])
    theta = np.deg2rad(angle)
    R = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta), np.cos(theta)]])
    pts = np.matmul(R, points.T).T
    return sg.Polygon(pts)

def polygonPoints(poly):
    points = []
    for vertex in list(poly.vertices):
        pt = []
        pt.append(vertex.x())
        pt.append(vertex.y())
        points.append(pt)
    return np.array(points)

def configSpaceGen(space, robot):
    configSpace = []
    pts = polygonPoints(robot)
    R = np.array([[-1, 0],
                  [0, -1]])
    robotPoints = np.matmul(R, pts.T).T
    for s in space:
        wall = polygonPoints(s)
        v = []
        for v1 in robotPoints:
            for v2 in wall:
                tmp = v1 + v2
                v.append(sg.Point2(tmp[0],tmp[1]))        
        hullPts = sg.convex_hull.graham_andrew(v)
        configSpace.append(sg.Polygon(hullPts))
    return configSpace

def isPointFree(space,point):
    free = True
    for s in space:
        if(s.oriented_side(point) == sg.Sign.POSITIVE ):
            free = False
            break
    return free

def getNextPoint(x,y,ang):
    ADJ_S = {0: (1, 0), 5: (11, 1), 10: (6, 1), 15: (4, 1), 20: (3, 1), 25: (2, 1), 30: (9, 5), 35: (3, 2), 40: (6, 5), 45: (1, 1),
         50: (5, 6), 55: (2, 3), 60: (5, 9), 65: (1, 2), 70: (1, 3), 75: (1, 4), 80: (1, 6), 85: (1, 11), 90: (0, 1)}
    ax, ay = ADJ_S[ang]
    return x + ax, y + ay

def isSegmentFree(space,p1,p2,angle):
    free = True
    x1 = p1.x()
    y1 = p1.y()
    num_samples = int(LA.norm((x1 - p2.x(), y1 - p2.y())))
    while(x1!=p2.x() and y1!=p2.y()):
        if(isPointFree(space,sg.Point2(x1,y1)) == False):
            free = False
            break
        theta = np.deg2rad(angle) 
        x1 = x1 + 1/num_samples * np.cos(theta)
        y1 = y1 + 1/num_samples * np.sin(theta)
    return free      
    
```

```python
robot = robotGenerator()
newSpace = configSpaceGen(env,robot)
draw(newSpace)
```

```python
robot = robotGenerator(45)
newSpace = configSpaceGen(env,robot)
draw(newSpace)
```

```python
cspace = {}
for ang in range(0, 91, 5):
    robot = robotGenerator(ang)
    space = configSpaceGen(env,robot)
    cspace[ang] = space
```

```python
def generateGraph(cspace):
    begin = time.time()
    graph = nx.Graph()
    for x in range(1, 801):
        for y in range(1, 301):
            for ang in range(0, 91, 5):
                pt = sg.Point2(x,y)
                # Add point to graph after rotation
                if ang != 90:
                    if (isPointFree(cspace[ang],pt) and isPointFree(cspace[ang+5],pt)) :
                        graph.add_edge((x, y, ang), (x, y, ang + 5), weight=0)
                # Add point to graph after translation
                x1, y1 = getNextPoint(x,y,ang)
                pt1 = sg.Point2(x1,y1)
                if(isSegmentFree(cspace[ang],pt,pt1,ang)):
                    if isPointFree(cspace[ang],pt) and isPointFree(cspace[ang],pt1) :
                        w = LA.norm((x1 - x, y1 - y))
                        graph.add_edge((x, y, ang), (x1, y1,ang), weight=w)                
    end = time.time()
    print(f'Graph took {end - begin:.4} seconds.')
    return graph
```

```python
graph = generateGraph(cspace)
```

```python
t0 = time.time()
start = (50, 50, 0)
target = (750, 50, 90)
length, shortest_path = nx.algorithms.bidirectional_dijkstra(graph, start, target)
t1 = time.time()
print(f'Dijkstra took {t1 - t0:.4} seconds.')
```

```python
print('Length of path: ', length)
seg = []
for i in range(len(shortest_path)):
    if i > 1:
       s1 = sg.Segment2(sg.Point2(shortest_path[i][0], shortest_path[i][1]), sg.Point2(shortest_path[i-1][0],shortest_path[i-1][1]))
       seg.append(s1)
draw(env)
draw(seg)
```

```python
from matplotlib import pyplot as plt
t = np.array(shortest_path)
start = (50, 50, 0)
target = (750, 50, 90)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(start[0], start[1], start[2], label='Start', s=200, c='blue')
ax.scatter(target[0], target[1], target[2], label='Target', s=200, c='black')

ax.plot(t[:, 0], t[:, 1], t[:, 2], c='red', label='Shortest Path')
plt.legend()
plt.show()
```

```python
def random_point(xmin=0, xmax=801, ymin=0, ymax=301, tint=5):
    rx = np.random.randint(xmin, xmax)
    ry = np.random.randint(ymin, ymax)
    rtheta = np.random.choice(list(range(0, 91, tint)))
    return (rx, ry, rtheta)
    
def generatePRMGraph(start, target, cspace, sampleNum=50):
    graph = nx.Graph()
    graph.add_node(start)
    graph.add_node(target)
    count = 0
    while count < sampleNum:
        arr = np.array(graph.nodes)
        pt = random_point()
        tr = cKDTree(arr[:, :2])
        d, i = tr.query([pt[0], pt[1]], k=min(15, arr.shape[0]))
        connected = False
        for n in arr[i]:
            if(isSegmentFree( cspace[n[2]], sg.Point2(pt[0],pt[1]) , sg.Point2(n[0],n[1]) , n[2])):
                flag1 = False
                flag2 = False
                if n[2] != pt[2]:
                    if (isPointFree(cspace[n[2]] , sg.Point2(pt[0],pt[1]))) and (isPointFree(cspace[pt[2]],sg.Point2(pt[0],pt[1]))) :
                        flag1 = True
                        graph.add_edge((pt[0], pt[1], pt[2]), (pt[0], pt[1], n[2]), weight=0)
                    else : 
                        flag1 = False
                    if not flag1:
                        continue
                if (isPointFree(cspace[n[2]],sg.Point2(n[0],n[1]))) and (isPointFree(cspace[n[2]],sg.Point2(pt[0],pt[1]))) :
                    w = LA.norm((pt[0] - n[0], pt[1] - n[1]))
                    graph.add_edge((pt[0], pt[1],n[2]),tuple(n), weight=w)
                    flag2 = True
                else :
                    flag2 = False
                connected |=flag2
        if connected:
            count += 1
        connected = False
    return graph 
```

```python
start = (50, 50, 0)
target = (750, 50, 0)
begin = time.time()
prm50 = generatePRMGraph(start, target, cspace, 100)
end = time.time()
print(f'Building this PRM took {end - begin:.4} seconds.')
```

```python
num_components = nx.components.number_connected_components(prm50)
print(f'The resulting graph has {num_components} connected components.')
```

```python
start = (50, 50, 0)
target = (750, 50, 0)
length, shortest_path = nx.algorithms.bidirectional_dijkstra(prm50, start, target)
```

```python
print('Length of path: ', length)
seg = []
for i in range(len(shortest_path)):
    if i > 1:
       s1 = sg.Segment2(sg.Point2(shortest_path[i][0], shortest_path[i][1]), sg.Point2(shortest_path[i-1][0],shortest_path[i-1][1]))
       seg.append(s1)
draw(env)
draw(seg)
```
